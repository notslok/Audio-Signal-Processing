```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

### *Loading audio files using librosa*

```python
scale_file = "/content/drive/MyDrive/Material_logs_shlok/Speech_Processing_Material/ASP/18 - Extracting Mel Spectrograms with Python/audio/scale.wav"
```

```python
ipd.Audio(scale_file)
```

```python
# load audio files with librosa
scale, sr = librosa.load(scale_file)
```

### *Mel filter banks*

```python
filter_banks = librosa.filters.mel(n_fft = 2048, sr = 22050, n_mels = 10) # n_mels->selecting number of mel banks  
```

```python
filter_banks.shape #(np of mel filter banks, nyqst frequency + 1)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks,
                         sr = sr,
                         x_axis = "linear"
                         )
plt.colorbar(format = "%+2.f")
plt.show() # shows the mel bank's spread over frequencies
```

### *Extracting Mel Spectrogram*

```python
mel_spectrogram = librosa.feature.melspectrogram(scale, sr = sr, n_fft = 2048, hop_length = 512, n_mels = 120)
```

```python
mel_spectrogram.shape
```

```python
scale.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
plt.figure(figsize=(25,10))

librosa.display.specshow(log_mel_spectrogram,
                         sr = sr,
                         x_axis = "time",
                         y_axis = "mel")
plt.colorbar(format = "%+2.f")
plt.show() # in this plot y-axis is made of discrete chunks 
```
