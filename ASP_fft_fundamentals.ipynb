```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load audio files in the player
audio_path = "/content/drive/MyDrive/Material_logs_shlok/Speech_Processing_Material/ASP/10 - Fourier Transform: The Intuition/audio/piano_c.wav"
ipd.Audio(audio_path)
```

```python
# load audio file
signal, sr = librosa.load(audio_path)
```

```python
# plot waveform
plt.figure(figsize=(18, 8))
librosa.display.waveplot(signal, sr = sr, alpha = 0.5)
plt.show()
```

```python
# deriving frequency domain spectrum using FFT
ft = sp.fft.fft(signal)
magnitude = np.absolute(ft)
frequency = np.linspace(0,sr,len(magnitude))
```

```python
# plotting the spectrum
plt.figure(figsize = (18,8))
plt.plot(frequency[:5000], magnitude[:5000])
plt.xlabel("Frequency(Hz)")
plt.ylabel("Magnitude")
plt.show()
```

```python
# zoom into waveform
samples = range(len(signal))
t = librosa.samples_to_time(samples, sr = sr)

plt.figure(figsize=(18,8))
plt.plot(t[10000:10400],signal[10000:10400])
# plt.plot(t,signal)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# crate a sinusoid

f = 523
phase = 0
phase_2 = 0.2

sin = 0.5*np.sin(2*np.pi*(f*t - phase))
sin_2 = 0.5*np.sin(2*np.pi*(f*t - phase_2))

plt.figure(figsize = (18,8))
plt.plot(t[10000:10400], sin[10000:10400], color="b")
plt.plot(t[10000:10400], sin_2[10000:10400], color="g")

plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Comparing Signal and Sinusoid

f = 523
phase = 0.55

sin = 0.1*np.sin(2*np.pi*(f*t - phase))

plt.figure(figsize = (18,8))
plt.plot(t[10000:10400], signal[10000:10400], color="b")
plt.plot(t[10000:10400], sin[10000:10400], color="g")

plt.fill_between(t[10000:10400], sin[10000:10400]*signal[10000:10400], color="y")

plt.ylabel("Amplitude")
plt.xlabel("Time(s)")
plt.show()
```

```python
# Superimposing pure tones

f = 1
t = np.linspace(0, 10, 10000)

sin1 = np.sin(2*np.pi * (f*t))
sin2 = 3*np.sin(2*np.pi * (1.5*f*t))
sin3 = 7*np.sin(2*np.pi * (6*f*t))


sin_complex_imposed = sin1 + sin2 + sin3

```

```python
plt.figure(figsize=(15, 10))

plt.subplot(4, 1, 1)
plt.plot(t, sin1)

plt.subplot(4, 1, 2)
plt.plot(t, sin2)

plt.subplot(4, 1, 3)
plt.plot(t, sin3)

plt.subplot(4, 1, 4)
plt.plot(t, sin_complex_imposed, color="r")

plt.xlabel("Time -->")
plt.ylabel("Amplitude -->")

plt.show()
```
